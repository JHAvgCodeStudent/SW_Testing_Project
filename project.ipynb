# Spectrum Based Fault Localization

```python
import math

#Tarantula
def tarantula(line_num_failed, num_total_fail, line_num_pass, num_total_pass):
    numerator = line_num_failed / num_total_fail
    denominator = numerator + (line_num_pass / num_total_pass)
    suspicion = numerator / denominator
    return suspicion

#sbi
def sbi(line_num_failed, line_num_pass):
    suspicion = line_num_failed / (line_num_failed + line_num_pass)
    return suspicion

#Jaccard
def jaccard(line_failed, line_pass, total_failed):
    suspicion = line_failed / (total_failed + line_pass)
    return suspicion

#Ochiai
def ochiai(line_failed, line_passed, total_failed):
    numerator = line_failed
    denominator = math.sqrt((total_failed *(line_passed + line_failed)))
    suspicion = numerator / denominator
    return suspicion
```
